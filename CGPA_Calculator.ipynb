CGPA Calculator

```python
def calculate_cgpa():
    print("CGPA Calculator")

    # 1. Select Scale
    while True:
        try:
            scale = int(input("Enter your scale (4 or 5): "))
            if scale in [4, 5]:
                break
            print("Please enter exactly 4 or 5.")
        except ValueError:
            print("Invalid input. Please enter a number (4 or 5).")

    # 2. Number of Courses
    while True:
        try:
            num_courses = int(input("How many courses did you take? "))
            if num_courses > 0:
                break
            print("Number of courses must be greater than 0.")
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    total_points = 0
    total_units = 0

    # 3. Course Entry Loop
    for i in range(1, num_courses + 1):
        print(f"\n[ Course {i} ]")

        # Retry loop for Units
        while True:
            try:
                unit = float(input(f"  Enter Credit Units for Course {i}: "))
                break # Exit the 'while' loop because input is valid
            except ValueError:
                print(f"  !! Invalid input for Units. Please try Course {i} again.")

        # Retry loop for Grade Point
        while True:
            try:
                gp = float(input(f"  Enter Grade Point for Course {i}: "))
                if 0 <= gp <= scale:
                    break # Exit the 'while' loop
                print(f"  !! Grade Point must be between 0 and {scale}.")
            except ValueError:
                print(f"  !! Invalid input for Grade Point. Please try Course {i} again.")

        total_points += (unit * gp)
        total_units += unit

    # 4. Results
    cgpa = total_points / total_units
    print("\n" + "="*30)
    print(f"FINAL CGPA: {cgpa:.2f} / {scale}.0")

    # Classification logic
    if scale == 5:
        if cgpa >= 4.5: remark = "First Class"
        elif cgpa >= 3.5: remark = "Second Class (Upper)"
        elif cgpa >= 2.4: remark = "Second Class (Lower)"
        elif cgpa >= 1.5: remark = "Third Class"
        else: remark = "Pass"
    else:
        if cgpa >= 3.5: remark = "Distinction"
        elif cgpa >= 3.0: remark = "Upper Credit"
        elif cgpa >= 2.5: remark = "Lower Credit"
        elif cgpa >= 2.0: remark = "Pass"
        else: remark = "Fail"

    print(f"REMARK: {remark}")
    print("="*30)

calculate_cgpa()
```
